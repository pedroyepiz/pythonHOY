Cuando se asigna un nuevo valor a una variable creada anteriormente, el valor de la variable será el nuevo valor almacenado, pero con respecto a los objetos (que son los que realmente almacenan los valores) pueden darse tres situaciones:

  + Que el objeto original sea inmutable y no pueda almacenar el nuevo valor. En este caso, la variable se asocia a un objeto distinto que contiene el nuevo valor.

  + Que el objeto original sea mutable, pero que no se modifique. En este caso, la variable se asocia a un objeto distinto que contiene el nuevo valor.

  + Que el objeto original sea mutable y pase a contener el nuevo valor. En este caso, la variable puede seguir asociada al mismo objeto.

En Python, cuando se asigna un valor a una variable, lo que realmente se está haciendo es asignar una referencia a un objeto en memoria. Cuando se modifica el valor de una variable, lo que realmente se está haciendo es cambiar la referencia de la variable al objeto en memoria.

Cuando se modifica una variable, se puede modificar el objeto al que apunta la referencia. Si el objeto es mutable, como una lista o un diccionario, se pueden modificar sus valores internos. Si el objeto es inmutable, como un número o una cadena, no se puede modificar su valor interno. En cambio, se puede crear un nuevo objeto y asignar la referencia a la variable modificada.

## DIFERENCIA DE MODIFICACION DE VARIABLE ENTRE C Y PYTHON
 En C, cuando se declara una variable, se le asigna un espacio de memoria en el que se almacenará su valor. Al modificar la variable, se actualiza directamente ese valor en ese espacio de memoria específico. Es como alquilar una casa: se tiene acceso a ella temporalmente y se puede hacer modificaciones directamente en ella.

En Python, por otro lado, cuando se declara una variable, lo que se almacena en memoria es una referencia a un objeto. Al modificar la variable, lo que se está haciendo es cambiar la referencia a un objeto diferente en memoria. Es como comprar una casa: se tiene acceso a ella a través de una dirección específica, pero si se desea cambiar la casa, se cambia la dirección y se accede a una casa diferente.

En resumen, en C las variables se modifican directamente en su espacio de memoria asignado, mientras que en Python se modifican las referencias a objetos en memoria.

## **En Python, la modificación de variables sigue estos pasos:**

**Asignación de variables:** Se asigna un valor a una variable usando el operador de asignación =.

**Creación de objeto:** Se crea un objeto que contiene el valor asignado.

**Referencia al objeto:** La variable hace referencia al objeto creado.

**Manipulación de objeto:** Se puede manipular el objeto a través de la variable asignada.

Es importante destacar que en Python las variables son referencias a objetos, lo que significa que cuando se modifica una variable, no se está modificando el valor de la variable en sí misma, sino el objeto al que hace referencia la variable.
